# Factorial


````python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        return "Factorial is not defined for negative numbers or non-integers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Test the function with some numbers
print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 0: {factorial(0)}")
print(f"Factorial of 10: {factorial(10)}")
print(f"Factorial of -3: {factorial(-3)}")
print(f"Factorial of 3.5: {factorial(3.5)}")
````

# Multiplication Table


````python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Multiplies from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Test the function with a number
display_multiplication_table(7)
print("\n") # Add a newline for better readability
display_multiplication_table(12)
````

# Odd & Even


````python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Test the function with some examples
print(check_odd_even(15))
print(check_odd_even(20))
print(check_odd_even(1))
print(check_odd_even(0))
````

# Palindrome

````python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Reverse the string
    reversed_str = num_str[::-1]
    # Compare the original and reversed strings
    if num_str == reversed_str:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Test the function with some examples
print(is_palindrome(121))
print(is_palindrome(12321))
print(is_palindrome(12345))
print(is_palindrome(9))
print(is_palindrome(11))
````
