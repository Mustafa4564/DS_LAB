#FACTORIAL

````python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        return "Factorial is not defined for negative numbers or non-integers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# New testing section: Prompt user for input
print("\n--- Interactive Factorial Test ---")
user_input_str = input("Enter a non-negative integer to calculate its factorial: ")

try:
    user_number = int(user_input_str)
    # Call the factorial function with user-provided number
    result = factorial(user_number)
    # Print the result
    print(f"The factorial of {user_number} is: {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
````

#MULTIPLICATION TABLE

````python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Multiplies from 1 to 10
        print(f"{number} x {i} = {number * i}")

# New testing section: Prompt user for input
print("\n--- Interactive Multiplication Table Test ---")
user_input_str = input("Enter a number to display its multiplication table: ")

try:
    user_number = int(user_input_str)
    # Call the display_multiplication_table function with user-provided number
    display_multiplication_table(user_number)
except ValueError:
    print("Invalid input. Please enter an integer.")
````

#ODD EVEN

````python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# New testing section: Prompt user for input
print("\n--- Interactive Odd/Even Test ---")
user_input_str = input("Enter an integer to check if it's odd or even: ")

try:
    user_number = int(user_input_str)
    # Call the check_odd_even function with user-provided number
    result = check_odd_even(user_number)
    # Print the result
    print(result)
except ValueError:
    print("Invalid input. Please enter an integer.")
````

#PALINDROME

````python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Reverse the string
    reversed_str = num_str[::-1]
    # Compare the original and reversed strings
    if num_str == reversed_str:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# New testing section: Prompt user for input
print("\n--- Interactive Palindrome Test ---")
user_input_str = input("Enter a number to check if it's a palindrome: ")

try:
    user_number = int(user_input_str)
    # Call the is_palindrome function with user-provided number
    result = is_palindrome(user_number)
    # Print the result
    print(result)
except ValueError:
    print("Invalid input. Please enter an integer.")
````
